```python
#Import pandas to scrape variable list tables
import pandas as pd
```

```python
# Urls and code to scrape tables
dem_url = 'https://wwwn.cdc.gov/Nchs/Nhanes/Search/variablelist.aspx?Component=Demographics&CycleBeginYear=2013'
exa_url = 'https://wwwn.cdc.gov/Nchs/Nhanes/Search/variablelist.aspx?Component=Examination&CycleBeginYear=2013'
lab_url = 'https://wwwn.cdc.gov/Nchs/Nhanes/Search/variablelist.aspx?Component=Laboratory&CycleBeginYear=2013'
diet_url = 'https://wwwn.cdc.gov/Nchs/Nhanes/Search/variablelist.aspx?Component=Dietary&CycleBeginYear=2013'
que_url = 'https://wwwn.cdc.gov/Nchs/Nhanes/Search/variablelist.aspx?Component=Questionnaire&CycleBeginYear=2013'
dem_info = pd.read_html(dem_url)[0]
exa_info = pd.read_html(exa_url)[0]
lab_info = pd.read_html(lab_url)[0]
diet_info = pd.read_html(diet_url)[0]
que_info = pd.read_html(que_url)[0]
```

```python
#Read in table to verify it worked
dem_info.head()
```

```python
#Save tables as CSVs
dem_info.to_csv('../data/CSVs/demographics_variable_list.csv', index=False)
exa_info.to_csv('../data/CSVs/examination_variable_list.csv', index=False)
diet_info.to_csv('../data/CSVs/dietary_variable_list.csv', index=False)
lab_info.to_csv('../data/CSVs/laboratory_variable_list.csv', index=False)
que_info.to_csv('../data/CSVs/questionnaire_variable_list.csv', index=False)
```
